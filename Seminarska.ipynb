```python
import pandas as pd

data=pd.read_csv("/content/drive/MyDrive/data.csv")   # read the whole data
```

```python
# Now i check, if for example the row for Sweden, for January month 2018 year is missing, I drop
# all the rows for that year, because I must have consistency when I work with time series.
```

```python
# So, if the number of rows for some year, for some country, for specific product (I check for each of products, years and countries)
# is less than twelve, that means the value for some month is missing, then drop all the rows for that product, in specific year, in specific country
```

```python
list_countires = ['Sweden', 'Albania', 'Slovenia', 'North Macedonia', 'Spain', 'Serbia']
list_foods = ['Apples', 'Bananas', 'Beans, dry', 'Cabbages', 'Carrots and turnips', 'Cauliflowers and broccoli', 'Cucumbers and gherkins', 'Grapes', 'Green garlic', 'Lemons and limes', 'Lettuce and chicory', 'Oats', 'Olives', 'Oranges', 'Peas, green', 'Mushrooms and truffles']
```

```python
for country in list_countires:
  list_foods = ['Apples', 'Bananas', 'Beans, dry', 'Cabbages', 'Carrots and turnips', 'Cauliflowers and broccoli', 'Cucumbers and gherkins', 'Grapes', 'Green garlic', 'Lemons and limes', 'Lettuce and chicory', 'Oats', 'Olives', 'Oranges', 'Peas, green', 'Mushrooms and truffles']

  for food in list_foods:

    if len(data.loc[(data['Area'] == country) & (data['Item'] == food) & (data['Year'] == 2018)]) < 12:
      lista_indeksi = data.index[(data['Area'] == country) & (data['Item'] == food) & (data['Year'] == 2018)].tolist()
      data = data.drop(index=lista_indeksi)

    if len(data.loc[(data['Area'] == country) & (data['Item'] == food) & (data['Year'] == 2019)]) < 12:
      lista_indeksi = data.index[(data['Area'] == country) & (data['Item'] == food) & (data['Year'] == 2019)].tolist()
      data = data.drop(index=lista_indeksi)

    if len(data.loc[(data['Area'] == country) & (data['Item'] == food) & (data['Year'] == 2020)]) < 12:
      lista_indeksi = data.index[(data['Area'] == country) & (data['Item'] == food) & (data['Year'] == 2020)].tolist()
      data = data.drop(index=lista_indeksi)

    if len(data.loc[(data['Area'] == country) & (data['Item'] == food) & (data['Year'] == 2021)]) < 12:
      lista_indeksi = data.index[(data['Area'] == country) & (data['Item'] == food) & (data['Year'] == 2021)].tolist()
      data = data.drop(index=lista_indeksi)
```

```python
# Now i convert Value column, from LCU (Local currency units) to Euro, for that countries which don't have
# euros as payment currency
```

```python
for index, row in data.iterrows():
  if row['Area'] == 'Albania':
    #print(data.loc[index]['Value'])
    #print(row['Value'] * 0.0094)
    data.at[index, 'Value'] = row['Value'] * 0.0094

  if row['Area'] == 'Sweden':
    data.at[index, 'Value'] = row['Value'] * 0.084

  if row['Area'] == 'North Macedonia':
    data.at[index, 'Value'] = row['Value'] * 0.016

  if row['Area'] == 'Serbia':
    data.at[index, 'Value'] = row['Value'] * 0.0085
```

```python
data = data.reset_index() # After drop some rows, I have to reset indexes
```

```python
# Now I make new DataFrame, where I have five columns
# The last three columns are prices for one month ago, two months ago etc.
```

```python
matrix = []
for i in range(len(data['Value'])-1,2,-1):

  month3=data['Value'][i-3]
  month2=data['Value'][i-2]
  month1=data['Value'][i-1]
  matrix.append([data['Year'][i], data['Area'][i], data['Item'][i], data['Value'][i],month1,month2,month3])

matrix


df=pd.DataFrame(matrix,columns=['Year', 'Country', 'Product', 'Predicted value','One month ago','Two months ago','Three months ago'])
```

```python
# Now i transform Country and Product columns to integer, because I need integer inputs for models
# Also I have to transform Year column with label encoder, because if I don't do it, it will use the years as numbers
# not the actual year
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['Year']=le.fit_transform(df['Year'])
```

```python
df['Country'] = le.fit_transform(df['Country'])
```

```python
df['Product'] = le.fit_transform(df['Product'])
```

```python
test = df[:int(len(df)*0.2)]
train = df[int(len(df)*0.2):]
```

```python
X_train = train.iloc[:, :3].join(train.iloc[:, -3:])
X_test = test.iloc[:, :3].join(test.iloc[:, -3:])
Y_train = train['Predicted value']
Y_test = test['Predicted value']
```

```python
X_test
```

```python
import numpy as np

X_train_array = X_train.to_numpy()
X_test_array = X_test.to_numpy()


X_train_reshaped = np.reshape(X_train_array, (X_train_array.shape[0], 1, X_train_array.shape[1]))
X_test_reshaped = np.reshape(X_test_array, (X_test_array.shape[0], 1, X_test_array.shape[1]))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import r2_score


model = Sequential()
model.add(LSTM(100, input_shape=(X_train_reshaped.shape[1], X_train_reshaped.shape[2]), activation='relu', return_sequences=True))
model.add(LSTM(64, activation='relu', return_sequences=True))
model.add(LSTM(32, activation='relu', return_sequences=True))
model.add(LSTM(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='linear'))

model.compile(optimizer='adam', loss='mean_squared_error')

model.fit(X_train_reshaped, Y_train, epochs=150, batch_size=16)
```

```python
y_pred = model.predict(X_test_reshaped)
from sklearn.metrics import r2_score
print(r2_score(Y_test,y_pred))
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(Y_test,y_pred))
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, Flatten, Dense

X_train_array = X_train.to_numpy()
X_test_array = X_test.to_numpy()
Y_train_array = Y_train.to_numpy()
Y_test_array = Y_test.to_numpy()

timesteps = 3
features = X_train_array.shape[1]

X_train_reshaped = np.array([X_train_array[i:i + timesteps] for i in range(len(X_train_array) - timesteps + 1)])
X_test_reshaped = np.array([X_test_array[i:i + timesteps] for i in range(len(X_test_array) - timesteps + 1)])

# Reshape the target data
Y_train_reshaped = Y_train_array[timesteps - 1:]
Y_test_reshaped = Y_test_array[timesteps - 1:]

# Define the CNN model
model = Sequential()
model.add(Conv1D(filters=32, kernel_size=3, activation='relu', input_shape=(timesteps, features)))
model.add(Conv1D(filters=64, kernel_size=3, activation='relu', padding='same'))
model.add(Conv1D(filters=128, kernel_size=3, activation='relu', padding='same'))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='linear'))


model.compile(optimizer='adam', loss='mean_squared_error')


model.fit(X_train_reshaped, Y_train_reshaped, epochs=150, batch_size=16, validation_split=0.2)


loss = model.evaluate(X_test_reshaped, Y_test_reshaped)
print(f"Test Loss: {loss}")


Y_pred = model.predict(X_test_reshaped)
mae = mean_absolute_error(Y_test_reshaped, Y_pred)
r2 = r2_score(Y_test_reshaped, Y_pred)

print(f"Mean Absolute Error: {mae}")
print(f"R-squared Score: {r2}")
```
